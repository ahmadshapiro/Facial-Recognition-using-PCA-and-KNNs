## 1- Data Preprocessing 

```python
import os
import seaborn as sns 
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from PIL import Image
from skimage.io import imread, imshow
from sklearn.metrics import accuracy_score
np.random.seed(42)
%matplotlib inline 
```

```python
#base folder where the data is 
base_dir = './archive/' 
```

```python
#Lets read one image and check the dimensions 
k = imread("./archive/s1/1.pgm", as_gray=True)
imshow(k);
```

```python
k
```

```python
print(k.shape) #as we can see the dimensions 
#so for flatten the dimensions we will need 112 * 92 feature 
num_features=k.shape[0]*k.shape[1]
print("Number of features = {}".format(num_features))
```

```python
#now we will have column names for our features 
col_names = ["f_"+str(x) for x in range(1,num_features+1)]
len(col_names)
#now let's add Label to the columns names 
col_names.append("Label")
```

```python
list_of_rows = []
```

```python
for _,subfolders,__ in os.walk(base_dir):
    for folder in subfolders:
        for _,_,files in os.walk(base_dir+folder):
            for image in files :
                label = folder.replace("s","")
                img_path=base_dir+folder+'/'+image
                image_file= list(imread(img_path,as_gray=True).flatten())
                image_file.append(label)
                k=dict()
                for key,pixel in zip(col_names,image_file):
                    k[key]=pixel
                list_of_rows.append(k)
```

```python
df=pd.DataFrame(data=list_of_rows,columns=col_names)
```

```python
df
```

```python
Y = df.Label
Y
```

```python
df.drop(columns=["Label"],inplace=True)
df
```

Let's take a random picture to check if the preprocessig was right

```python
randomimage=np.random.randint(1,400)
plt.imshow(df.iloc[randomimage,0:].values.reshape(112,92),cmap="gray");
```

### Now we should scale our data 

```python
df=df/255.
df
```

```python
randomimage=np.random.randint(1,400)
plt.imshow(df.iloc[randomimage,0:].values.reshape(112,92),cmap='gray');
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(df,Y,stratify=Y,test_size=0.5) 
```

```python
from collections import Counter
```

```python
Counter(y_train) , Counter(y_test)
```

Now we can see that each person has 5 pictures per training and testing data sets 

### 2- PCA 

```python
cols = ['Alpha','Explained Variance','Number of Components','Accuracy']
alphas = np.arange(0.1,0.97,0.05).round(3)
results_1=pd.DataFrame(columns=cols)

lists=[]
for alpha in alphas :
    classifier= KNeighborsClassifier(n_neighbors=1)
    pca = PCA(n_components=alpha)
    X_train_trans = pca.fit_transform(X_train)
    X_test_trans= pca.transform(X_test)
    classifier.fit(X_train_trans,y_train)
    y_hat=classifier.predict(X_test_trans)
    acc = accuracy_score(y_test,y_hat)
    row = {'Alpha': alpha
           ,'Explained Variance':sum(pca.explained_variance_ratio_).round(3)
           ,'Number of Components': pca.n_components_
           ,'Accuracy': acc }
    lists.append(row)
results_1=results_1.append(lists)
results_1
```

```python
from matplotlib import style
style.use('ggplot')
```

```python
plt.figure(figsize=(10,5))
xi = list(range(len(results_1.Alpha[-4:])))
plt.plot(xi,results_1.Accuracy[-4:],marker="o");
plt.xticks(xi,results_1.Alpha[-4:]);
plt.title("Accuracy VS Explained Variance",fontsize=20);
plt.xlabel("Variance",fontsize=10);
plt.ylabel("Accuracy",fontsize=10);
```

Given the range of alpha above ranging from 0.8 to 0.95 we can't really spot the relation based only on 4 values of alpha , so we can plot the whole data to spot the trend 

```python
plt.figure(figsize=(20,10))
xi = list(range(len(results_1.Alpha)))
sizes=list(results_1['Number of Components'].values*10)
plt.plot(xi,results_1.Accuracy,color='black',alpha=0.3);
plt.scatter(xi,results_1.Accuracy,marker="o",s=sizes,color='red');
plt.xticks(xi,results_1.Alpha,fontsize=12);
plt.yticks(fontsize=12)
plt.title("Accuracy VS Explained Variance VS Components Number (Point Size)",fontsize=20);
plt.xlabel("Variance",fontsize=15);
plt.ylabel("Accuracy",fontsize=15);
```

```python
results_1
```

As we can see above that we were successfull spotting the trend in the data as follows 
<br>
1. As the explained variance get bigger the accuracy tends to increase , but till we reach the a bigger variance around 0.7 the accuracy reaches a steady state , but we cannot infere any relationship now since it's based on only 1 Neigherest Neighbor in the KNN Classifier, So we can tune the classifier to select the best K from range(1,20) and the best alpha is 0.95 because it reduced the number of featrues for 10305 to 116 variance explaining variables.
</br>
<br>
2. As the explained variance increases , the number of the PCs also increase. 

```python
pca = PCA(n_components=0.95)
pca.fit(X_train)
fig, axes = plt.subplots(2,10,figsize=(10,5),
 subplot_kw={'xticks':[], 'yticks':[]},
 gridspec_kw=dict(hspace=0.01, wspace=0.01))
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(112,92),cmap='gray')
```

### 3-Tunning The K-NN Classifier 

```python
from sklearn.model_selection import GridSearchCV
```

```python
pca = PCA(n_components=0.95)
X_train_trans = pca.fit_transform(X_train)
X_test_trans= pca.transform(X_test)
```

```python
metrics = ['euclidean','manhattan','minkowski', 'chebyshev'] 
neighbors = np.arange(1, 20)
leaf_size = np.arange(1,50)
p = [1,2]
weights= ['uniform','distance']
algorithm=['auto', 'ball_tree', 'kd_tree', 'brute']
param_grid  = dict(metric=metrics,
                   n_neighbors=neighbors,
                   leaf_size = np.arange(1,50),
                   weights=weights,
                   p=p,
                   algorithm=algorithm)
```

```python
knn=KNeighborsClassifier()
```

```python
grid_search = GridSearchCV(knn, param_grid,scoring='accuracy', refit=True)
grid_search.fit(X_train_trans, y_train)
```

```python
knn =grid_search.best_estimator_
```

```python
knn.get_params()
```

We can see from above that the best classifier having leaf size of 1 and K = 1

```python
grid_search.best_estimator_.score(X_test_trans,y_test)
```
